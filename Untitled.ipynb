```python
region_profit=df.groupby("Region")["Profit"].sum().sort_values(ascending=False)
state_profit=df.groupby("State")["Profit"].sum().sort_values(ascending=False).head(10)
country_profit=df.groupby("Country")["Profit"].sum().sort_values(ascending=False)
cateogry_profit=df.groupby("Category")["Profit"].sum().sort_values(ascending=False)
sub_category=df.groupby("Sub-Category")["Profit"].sum().sort_values(ascending=False)
segment=df.groupby("Segment")["Profit"].sum().sort_values(ascending=False)
productprofit=df.groupby("Product Name")["Profit"].sum().sort_values(ascending=False).head(10)
```

```python
display(region_profit)
display(state_profit)
display(country_profit)
display(cateogry_profit)
display(sub_category)
display(segment)
display(productprofit)
```

```python
plt.figure(figsize=(8, 5))
region_profit.plot(kind='bar')
plt.title('Total Profit by Region')
plt.xlabel('Region')
plt.ylabel('Total Profit')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(cateogry_profit, labels=cateogry_profit.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Proportion of Total Profit by Category')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
state_profit.plot(kind='bar')
plt.title('Top 10 States by Total Profit')
plt.xlabel('State')
plt.ylabel('Total Profit')
plt.xticks(rotation=45, ha='right') # Rotate labels for better readability
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Calculate Profit Margin (Profit / Sales * 100)
# Handle cases where Sales is 0 to avoid division by zero
df['Profit Margin'] = (df['Profit'] / df['Sales']) * 100
df['Profit Margin'] = df['Profit Margin'].replace([np.inf, -np.inf], np.nan) # Replace inf values with NaN
df['Profit Margin'] = df['Profit Margin'].fillna(0) # Fill NaN values (where sales was 0) with 0

display(df[['Sales', 'Profit', 'Profit Margin']].head())
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Profit Margin'], bins=50, kde=True)
plt.title('Distribution of Profit Margin')
plt.xlabel('Profit Margin (%)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
df['Order_Year_Month'] = df['Order Date'].dt.strftime('%Y-%m')
display(df[['Order Date', 'Order Year', 'Order Month', 'Order_Year_Month']].head())
```

```python
monthly_profit = df.groupby('Order_Year_Month')['Profit'].sum()
display(monthly_profit.head())
```

```python
plt.figure(figsize=(12, 6))
monthly_profit.plot(kind='line')
plt.title('Total Profit Over Time')
plt.xlabel('Month-Year')
plt.ylabel('Total Profit')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()
plt.show()
```
